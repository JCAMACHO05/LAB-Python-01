```python
inventario = [
    {"nombre": "Pan", "precio": 1500, "stock": 30},
    {"nombre": "Leche", "precio": 3500, "stock": 15},
    {"nombre": "Café", "precio": 8000, "stock": 10},
]

# Nombre primer producto
print(inventario[0]["nombre"])

# Agregar producto
inventario.append({"nombre": "Arroz", "precio": 2500, "stock": 20})

# Disminuir stock leche
for p in inventario:
    if p["nombre"] == "Leche":
        p["stock"] -= 2

# Valor total
total = sum(p["precio"] * p["stock"] for p in inventario)

# Stock menor a 10
bajos = [p for p in inventario if p["stock"] < 10]

# Aumentar precio 10%
for p in inventario:
    p["precio"] = round(p["precio"] * 1.10, 2)

# Buscar producto
buscar = input("Buscar producto: ")
encontrado = None

for p in inventario:
    if p["nombre"].lower() == buscar.lower():
        encontrado = p

print("Encontrado:", encontrado if encontrado else "No existe")

# Eliminar stock 0
inventario = [p for p in inventario if p["stock"] > 0]

# Ordenar por precio
inventario.sort(key=lambda x: x["precio"])

print("Inventario:", inventario)
```

```python
biblioteca = {
    "libros": [
        {"titulo": "1984", "autor": "George Orwell", "prestado": False},
        {"titulo": "El Principito", "autor": "Antoine de Saint-Exupéry", "prestado": True},
    ],
    "usuarios": ["Ana", "Luis", "María"]
}

prestados = sum(1 for l in biblioteca["libros"] if l["prestado"])
disponibles = len(biblioteca["libros"]) - prestados

print("Prestados:", prestados)
print("Disponibles:", disponibles)

# Agregar libro
biblioteca["libros"].append({"titulo": "Dune", "autor": "Frank Herbert", "prestado": False})

# Marcar 1984
for l in biblioteca["libros"]:
    if l["titulo"] == "1984":
        l["prestado"] = True

print(biblioteca)
```
