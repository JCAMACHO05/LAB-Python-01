```python

import requests

url = "https://jsonplaceholder.typicode.com/users"
usuarios = requests.get(url).json()

for u in usuarios:
    if u["email"].endswith(".biz"):
        print(u["name"])
```

```python
import requests

nombre = input("Pokémon: ").lower()
url = f"https://pokeapi.co/api/v2/pokemon/{nombre}"
res = requests.get(url)

if res.status_code == 200:
    datos = res.json()

    tipos = [t["type"]["name"] for t in datos["types"]]
    habilidades = [h["ability"]["name"] for h in datos["abilities"]]

    print("Tipos:", tipos)

    if len(tipos) > 1:
        print("Pokémon de tipo mixto")

    if datos["weight"] > 100:
        print("Pokémon de gran tamaño")
    else:
        print("Pokémon ligero")

    print("Habilidades:")
    for h in habilidades:
        print("-", h)

    # Extra: experiencia base
    print("Experiencia base:", datos["base_experience"])

else:
    print("No encontrado")
```
